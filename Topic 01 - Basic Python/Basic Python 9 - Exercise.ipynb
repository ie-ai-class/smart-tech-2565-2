````python
import numpy as np
import pandas as pd
num = 7
array = np.arange(1,num+1)
cols = [f"Col{i}" for i in array]
df = pd.DataFrame(columns=cols)
for i in array:
    df.loc[f'Row{i}'] = np.roll(array,-(i-1))
````

````python
html = df.to_html()
print(html)
````
